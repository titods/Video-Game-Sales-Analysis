```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

# Load dataset

```python
vgsales_2016 = pd.read_csv('dataset\Video Games Sales Dataset\Video_Games_Sales_as_at_22_Dec_2016.csv')
vgsales_2016
```

```python
vgsales_PS4 = pd.read_csv('dataset\Video Games Sales Dataset\PS4_GamesSales.csv')
vgsales_PS4
```

# Data Cleaning

## Select Columns

Take only selected columns for analysis such as **game name**, **game platform**, **release year**, **genre**, **publisher**, sales in **NA**, **EU**, **JP**, **Other**, and **global**.

*Note: Sales is in million number of copies*.

```python
# select columns in 'vgsales_2016'
vgsales_2016 = vgsales_2016[['Name', 'Platform', 'Year_of_Release', 'Genre', 'Publisher', 'NA_Sales', 
                            'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']]
vgsales_2016
```

```python
# create 'Platform' column
vgsales_PS4['Platform'] = 'PS4'

# rename the columns
vgsales_PS4.columns = ['Name', 'Year_of_Release', 'Genre', 'Publisher', 'NA_Sales', 
                       'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales', 'Platform']

# rearrange the columns order
vgsales_PS4 = vgsales_PS4[['Name', 'Platform', 'Year_of_Release', 'Genre', 'Publisher', 'NA_Sales', 
                            'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']]
vgsales_PS4
```

```python
# create 'Platform' column
vgsales_XOne['Platform'] = 'XOne'

# drop 'Pos' column
vgsales_XOne.drop('Pos', axis = 1, inplace = True)

# rename the columns
vgsales_XOne.columns = ['Name', 'Year_of_Release', 'Genre', 'Publisher', 'NA_Sales', 
                       'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales', 'Platform']

# rearrange the columns order
vgsales_XOne = vgsales_XOne[['Name', 'Platform', 'Year_of_Release', 'Genre', 'Publisher', 'NA_Sales', 
                            'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']]
vgsales_XOne
```

## Remove missing value

remove any missing values since we can't impute or predict it appropriately.

```python
# check missing values in 'vgsales_2016'
print('missing value in "vgsales_2016"')
print(vgsales_2016.isnull().sum())
```

```python
# check missing values in 'vgsales_PS4'
print('missing value in "vgsales_PS4"')
print(vgsales_PS4.isnull().sum())
```

```python
# check missing values in 'vgsales_XOne'
print('missing value in "vgsales_XOne"')
print(vgsales_XOne.isnull().sum())
```

```python
# remove any missing values in dataset
vgsales_2016 = vgsales_2016.dropna()
vgsales_PS4 = vgsales_PS4.dropna()
vgsales_XOne = vgsales_XOne.dropna()

# cross-check missing values
print(vgsales_2016.isnull().sum().sum())
print(vgsales_PS4.isnull().sum().sum())
print(vgsales_XOne.isnull().sum().sum())
```

## Change data types

`Year_of_Release` should be `object` data types

```python
# change data types in Year_of_Release column
vgsales_2016['Year_of_Release'] = vgsales_2016['Year_of_Release'].astype('int64').astype('object')
vgsales_PS4['Year_of_Release'] = vgsales_PS4['Year_of_Release'].astype('int64').astype('object')
vgsales_XOne['Year_of_Release'] = vgsales_XOne['Year_of_Release'].astype('int64').astype('object')
```

## Concatenate the dataset

Concatenate `vgsales_2016`, `vgsales_PS4`, and `vgsales_XOne`.

```python
# concatenate the dataset
vgsales = pd.concat([vgsales_2016, vgsales_PS4, vgsales_XOne], axis = 0).reset_index(drop = True)
```

## Remove duplicate observations

```python
# check duplicate observations in first 10 rows
vgsales[vgsales.duplicated(subset = ['Name','Platform'], keep = False)].sort_values(by = ['Name', 'Platform']).head(10)
```

```python
# check duplicate observations in last 10 rows
vgsales[vgsales.duplicated(subset = ['Name','Platform'], keep = False)].sort_values(by = ['Name', 'Platform']).tail(10)
```

```python
# number of duplicate games
print('total duplicates:', vgsales[vgsales.duplicated(subset = ['Name','Platform'], keep = False)].sort_values(by = ['Name', 'Platform']).shape[0]//2)
```

```python
# drop duplicates
vgsales = vgsales.drop_duplicates(subset = ['Name','Platform'], keep = 'first')
vgsales = vgsales.reset_index(drop = True)
```

## Sanity check the columns

```python
# describe the object data type column
display(vgsales.describe(include = 'object'))

# describe the non-object data type column
display(vgsales.describe(exclude = 'object'))
```

```python
# check unique values in 'Platform', 'Year_of_Release', 'Genre', 'Publisher'
for col in [column for column in vgsales.columns if column in ['Platform', 'Year_of_Release', 'Genre', 'Publisher']]:
    temp = vgsales[col].unique()
    temp.sort()
    print('unique values in "{}":\n{}'.format(col, temp))
    print()
```

Insights:
- `Platform` column could be grouped into group of console product. (i.e. Nintendo, Play Station, etc) 
    - &rarr; create new colum called `Console_Product`.
- `Publisher` column has too many unique values and some of them have a identical meaning with different named values.
    - &rarr; replace the identical value names with only one unique name.
    - &rarr; create new column called `Filtered_Publisher` to filter `Publisher` who has released more than 100 games.
- There are some zeros million number of copies in `Global_Sales`. It could be removed since doesn't hold any information for analysis.
    - &rarr; create new column called `Global_Sales (zeros)`.

```python
# create group of console product
nintendo = ['3DS', 'DS', 'GB', 'GBA', 'GC', 'N64', 'NES', 'SNES', 'Wii', 'WiiU']
playstation = ['PS', 'PS2', 'PS3', 'PS4', 'PSP', 'PSV']
sega = ['DC', 'GEN', 'GG', 'SAT', 'SCD']
atari = ['2600']
_3DO = ['3DO']
NG = ['NG']
PC = ['PC']
NEC  = ['PCFX', 'TG16']
bandai = ['WS']
xbox = ['X360', 'XB', 'XOne']
vgsales['Console_Product'] = vgsales['Platform'].apply(lambda x: 'Nintendo' if x in nintendo else (
                                                                 'Playstation' if x in playstation else (
                                                                 'Sega' if x in sega else (
                                                                 'Atari' if x in atari else (
                                                                 '3DO' if x in _3DO else (
                                                                 'Neo Geo' if x in NG else (
                                                                 'PC' if x in PC else (
                                                                 'NEC' if x in NEC else (
                                                                 'Bandai' if x in bandai else (
                                                                 'Xbox')
                                                                 ))))))))
                                                      )
```

```python
# replace the identical value names with only one unique name
vgsales['Publisher'] = vgsales['Publisher'].replace({'Activision Value':'Activision',
                                                     'Bandai Namco Games':'Bandai Namco Entertainment',
                                                     'Big Ben Interactive':'Bigben Interactive',
                                                     'Codemasters Online':'Codemasters',
                                                     'EA Sports':'Electronic Arts',
                                                     'EA Games':'Electronic Arts',
                                                     'FuRyu Corporation':'FuRyu',
                                                     'GameMill':'GameMill Entertainment',
                                                     'Hudson Entertainment':'Hudson Soft',
                                                     'Idea Factory':'Idea Factory International',
                                                     'Interchannel-Holon':'Interchannel',
                                                     'Interplay Productions':'Interplay',
                                                     'Kadokawa Shoten':'Kadokawa Games',
                                                     'Marvelous Interactive':'Marvelous Entertainment',
                                                     'Marvelous Games':'Marvelous Entertainment',
                                                     'Maximum Family Games':'Maximum Games',
                                                     'Microsoft Studios':'Microsoft Game Studios',
                                                     'Milestone S.r.l':'Milestone',
                                                     'Milestone S.r.l.':'Milestone',
                                                     'NEC Interchannel':'NEC',
                                                     'Namco Bandai Games':'Bandai Namco Entertainment',
                                                     'Paon':'Paon Corporation',
                                                     'Paradox Development':'Paradox Interactive',
                                                     'Paradox Interactiveï¿½':'Paradox Interactive',
                                                     'Rebellion':'Rebellion Developments',
                                                     'SNK':'SNK Playmore',
                                                     'Sony Computer Entertainment':'Sony Interactive Entertainment',
                                                     'Sony Computer Entertainment America':'Sony Interactive Entertainment',
                                                     'Sony Computer Entertainment Europe':'Sony Interactive Entertainment',
                                                     'Sony Online Entertainment':'Daybreak Game Company',
                                                     'Sony Oznline Entertainment':'Daybreak Game Company',
                                                     'Spike Chunsoft':'Spike',
                                                     'Square':'Square Enix',
                                                     'Square Enix ':'Square Enix',
                                                     'Square EA':'Square Enix',
                                                     'SquareSoft':'Square Enix',
                                                     'THQ':'THQ Nordic',
                                                     'Ubisoft Annecy':'Ubisoft',
                                                     'Valve Software':'Valve'})
```

```python
# filter 'Publisher' for less than 100 games to be 'Other'
Filtered_Publisher = vgsales['Publisher'].value_counts()[vgsales['Publisher'].value_counts() >= 100].index
vgsales['Filtered_Publisher'] = vgsales['Publisher'].apply(lambda x: x if x in Filtered_Publisher else 'Other')
```

```python
# create zeros global sales
vgsales['Global_Sales (zeros)'] = vgsales['Global_Sales'] == 0
```
